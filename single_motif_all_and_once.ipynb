```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import random


#motif equation
def motif9(x,t):
    
    A =x[0]
    E =x[1]
    
    dA_dt=k1+k2-k3*A*k8/(k8+E)
    dE_dt=k6+k4*A/(k7+A)-k5*E
    
    return [dA_dt, dE_dt]

def motif10(x,t):
    
    A =x[0]
    E =x[1]
    
    dA_dt=k1+k2-k3*A*E/(k7+E)
    dE_dt=k6+k4-k5*E*A/(k8+A)
    
    return [dA_dt, dE_dt]

def motif11(x,t):
    
    A =x[0]
    E =x[1]
    
    dA_dt=k1*E/(k7+E)+k2-k3*A
    dE_dt=k6+k4*A/(k8+A)-k5*E
    
    return [dA_dt, dE_dt]

def motif12(x,t):
    
    A =x[0]
    E =x[1]
    
    dA_dt=k1+k2*k7/(k7+E)-k3*A
    dE_dt=k6+k4-k5*E*A/(k8+A)
    
    return [dA_dt, dE_dt]

def motif13(x,t):
    
    A =x[0]
    E =x[1]
    
    dA_dt=k1+k2-k3*A*k7/(k7+E)
    dE_dt=k6+k4-k5*E*k8/(k8+A)
    
    return [dA_dt, dE_dt]

def motif14(x,t):
    
    A =x[0]
    E =x[1]
    
    dA_dt=k1+k2-k3*A*E/(k8+E)
    dE_dt=k6+k4*k7/(k7+A)-k5*E
    
    return [dA_dt, dE_dt]

def motif15(x,t):
    
    A =x[0]
    E =x[1]
    
    dA_dt=k1+k2*E/(k7+E)-k3*A
    dE_dt=k6+k4-k5*E*k8/(k8+A)
    
    return [dA_dt, dE_dt]


def motif16(x,t):
    
    A =x[0]
    E =x[1]
    
    dA_dt=k1+k2*k7/(k7+E)-k3*A
    dE_dt=k6+k4*k8/(k8+A)-k5*E
    
    return [dA_dt, dE_dt]

#assign parameter(constant rate)
num=[0,1,5,10,100]
k1=random.choice(num)
k2=random.choice(num)
k3=random.choice(num)
k4=random.choice(num)
k5=random.choice(num)
k6=random.choice(num)
k7=random.choice(num)
k8=random.choice(num)

#initial concentratino of A, E
x0=[0.1,0.1]
#time

t= np.linspace(0,100,100)
x= [odeint(motif9,x0,t),odeint(motif10,x0,t),odeint(motif11,x0,t),odeint(motif12,x0,t),odeint(motif13,x0,t),odeint(motif14,x0,t),odeint(motif15,x0,t),odeint(motif16,x0,t)] #each motifs
y= ["motif9","motif10","motif11","motif12","motif13","motif14","motif15","motif16"]
i=0
for i in range(8):
  x[i] #type in motif which you want to use.
  A = x[i][:,0]
  E = x[i][:,1]
  print('k1:{},k2:{},k3:{},k4:{},k5:{},k6:{},k7:{},k8:{}'.format(k1,k2,k3,k4,k5,k6,k7,k8))
  #log scale.
  plt.semilogy(t,A, label="A")
  plt.semilogy(t,E, label="E")
  plt.legend()
  plt.title(y[i])
  plt.savefig("plot.pdf")
  plt.grid()
  plt.show()
  i+=1
```
